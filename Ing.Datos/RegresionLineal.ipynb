```python
import pandas as pd
from google.colab import drive
import numpy as np
import pandas as pd
import matplotlib as plt

from sklearn.linear_model import LinearRegression

drive.mount('/content/drive')
ruta = "/content/drive/MyDrive/Lasallista/tercer parcial/Datos.csv"

datos = pd.read_csv(ruta, sep=";")

datos
```

```python
datos["PIB anual"] = datos["PIB anual"].str.replace("M€", " ")
datos["Salario Minimo"] = datos["Salario Minimo"].str.replace(",00", " ")

datos
```

```python
datos["PIB anual"] = datos["PIB anual"].str.replace(",", ".")
datos["CO2"] = datos["CO2"].str.replace(",", ".")
datos["Población"] = datos["Población"].str.replace(".", "")
datos
```

```python
#Limpieza de datos

datos["PIB anual"] = datos["PIB anual"].fillna(0)

#variable dependiente: Salario mínimo; variables independientes: PIB anual,
#población y CO2

datos
```

```python
datos = datos.astype({'PIB anual': float, 'CO2':float, 'Salario Minimo':float, 'Población':int} )
datos
```

```python
#Matriz de correlación
print (datos.corr())
```

```python
#Modelo de regresión con 1 variable
import statsmodels.formula.api as smf
from patsy.builtins import *

r1 = smf.ols(formula = 'Q("Salario Minimo")~Población', data=datos).fit()

r1.summary()
```

```python
#Modelo de regresión con 2 variables
import statsmodels.formula.api as smf
from patsy.builtins import *

r2 = smf.ols(formula = 'Q("Salario Minimo")~Población+Año', data=datos).fit()

r2.summary()
```

```python
#Modelo de regresión con 3 variables
import statsmodels.formula.api as smf
from patsy.builtins import *

r3 = smf.ols(formula = 'Q("Salario Minimo")~Población+Q("PIB anual")+Año', data=datos).fit()

r3.summary()
```
