```python
import requests
import bs4

bs4.__version__
```

```python
respuesta = requests.get('https://www.enter.co/')

soup = bs4.BeautifulSoup(respuesta.text, 'html.parser')
```

```python
NoticiasUno = soup.select('div.entry-header h2.entry-title.h2 a')
for titulo in NoticiasUno:
  print(titulo.contents[0])
```

```python
NoticiasDos = soup.select('div.entry-content p')
for descripcion in NoticiasDos:
  print(descripcion.contents[0])
```

```python
NoticiasTres = soup.select('div.col-lg-6.col-md-6.col-sm-6.col-xs-12 div.entry-header div.entry-meta div.meta-item.meta-date span.updated.meta-icon')
for i in NoticiasTres:
  print(i.contents[0])
```

```python
imprimirTexto = {}

for i in range(0, len(NoticiasUno)):
  imprimirTexto[NoticiasUno[i].contents[0]] = NoticiasDos[i].contents[0] , NoticiasTres[i].contents[0]

print(imprimirTexto)
```
